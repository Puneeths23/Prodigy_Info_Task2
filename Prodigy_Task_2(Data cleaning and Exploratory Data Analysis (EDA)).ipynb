```python
data.dropna(inplace=True)
data_1=data.drop_duplicates()
data_1.info
```

```python
for f in data.columns:
    print(f,data[f].nunique())
```

```python
pclass_names={
1:'First Class',
2:'Second Class',
3:'Third Class'
}
data['Pclass name']=data['Pclass'].map(pclass_names)
data.head()
```

```python
data['Family']=data["SibSp"]+data["Parch"]
data.head()
```

```python
plt.subplots(figsize=(9,9))
sns.countplot(x='Sex',data=data)
plt.title("Distribution of Passenger's by Sex")
plt.show()
```

```python
sns.countplot(data=data, x="Pclass name", hue='Sex', palette='Set2')
plt.title("Distribution of passenger's classes")
```

```python
data.describe().T
```

```python
cont_val=data[['PassengerId','Survived','Pclass','Age','SibSp','Parch','Fare','Family']]
sns.set_style('darkgrid')
sns.set_palette('pastel')
sns.heatmap(cont_val.corr(),cmap='viridis',annot=True)
```

```python
sns.set_style('white')
sns.displot(data=data, x="Age", bins=15, hue='Survived', palette='Dark2', hue_order=[1, 0])
plt.title("Distribution by age")
```

```python
plt.subplots(figsize=(9,9))
sns.countplot(x='Family',data=data)
plt.title("Distribution of passenger's family members")
plt.show()
```
